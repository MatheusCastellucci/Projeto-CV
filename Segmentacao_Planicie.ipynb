```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
dicionario = {
    'P1': 'FFVI\\Maps\\planicie.jpg',
    'P2': 'FFVI\\Maps\\planicie2.jpg',
    'P3': 'FFVI\\Maps\\planicie3.jpg',
}

escolha = 'P1' 
original_image = Image.open(dicionario[escolha])
cv2_img = np.array(original_image)
plt.imshow(cv2_img)
plt.axis('off')
plt.title('Imagem Original')
```

```python
blurred_img = cv2.GaussianBlur(cv2_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)
blurred_img = cv2.GaussianBlur(blurred_img, (11, 11), 0)

plt.figure()
plt.imshow(blurred_img)
plt.axis('off')
plt.title('Imagem com Blur')
```

```python
#agora com o blur aplicado, vamos tentar transformar a imagem em preto e branco
gray = cv2.cvtColor(blurred_img, cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.title('Imagem em Preto e Branco')
```

```python
# agora vamos tentar aplicar um threshold para tentar separar as cores
_, thresh = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)
plt.figure()
plt.imshow(thresh, cmap='gray')
plt.axis('off')
plt.title('Imagem com Threshold')
```

```python
# agora vamos erodir a imagem para tentar tirar as machas brancas pequenas
kernel = np.ones((9,9), np.uint8)
eroded = cv2.erode(thresh, kernel, iterations=1)

plt.figure()
plt.imshow(eroded, cmap='gray')
plt.axis('off')
plt.title('Imagem Erodida')
```

```python
#agora colocar as manchas pretas na imagem orioginal
mask = cv2.cvtColor(eroded, cv2.COLOR_GRAY2BGR)
masked = cv2.bitwise_and(cv2_img, mask)
plt.figure()
plt.imshow(masked)
plt.axis('off')
plt.title('Imagem Final')
```
